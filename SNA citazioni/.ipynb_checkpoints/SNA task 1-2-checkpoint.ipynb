Arxiv HEP-TH (high energy physics theory) citation graph is from arXiv and covers all the citations. Edges from u to v indicate that a paper u cited another paper v. If a paper cites, or is cited by, a paper outside the dataset, the graph does not contain any information about this. The data is of the papers in the period from January 1993 to April 2003.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import cdlib
from cdlib import algorithms, viz, LifeCycle
```

```python
input_file = 'cit-HepTh.edges'
output_file = 'cit-HepTh_2.edges'
 
with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for line in infile:
        columns = line.split()
        if len(columns) > 2:
            del columns[2]
        outfile.write(' '.join(columns) + '\n')
```

```python
def read_custom_edgelist(file_path):
    G = nx.Graph()
    with open(file_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2:
                u, v = parts[:2]
                G.add_edge(u, v)
    return G

G = read_custom_edgelist('cit-HepTh_2.edges')
```

```python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
```

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = np.array(degree_sequence)
bins = np.arange(min(degree_count), max(degree_count) + 1)

plt.hist(degree_count, bins=bins, density=True, histtype='step', cumulative=-1)
plt.title("Degree Distribution (CCDF)")
plt.xlabel("Degree")
plt.ylabel("CCDF")
plt.show()
```

```python
density = nx.density(G)
density
```

```python
clustering_coeff = nx.average_clustering(G)
clustering_coeff
```

```python
degree_centrality = nx.degree_centrality(G)
degree_centrality
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
betweenness_centrality
```

```python
closeness_centrality = nx.closeness_centrality(G)
closeness_centrality
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G)
eigenvector_centrality
```

```python
assortativity = nx.degree_assortativity_coefficient(G)
assortativity
```

```python
plt.figure(figsize=(10, 7))
plt.hist(list(degree_centrality.values()), bins=30)
plt.title("Degree Centrality Distribution")
plt.xlabel("Degree Centrality")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)
nx.draw(G, pos, node_color=list(degree_centrality.values()), node_size=10, cmap=plt.cm.viridis, with_labels=False)
plt.title("Graph Visualization with Node Color Indicating Degree Centrality")
plt.show()
```
